```python
save_base = '/home/haotian/molecules_confs/Protein_test/SurfGen/data/crossdock2020_surface_8'
crossdock2020 = '/home/haotian/molecules_confs/Protein_test/Res2mol/data/crossdocked_pocket10'
dirs = glob(crossdock2020+'/*')
fail_list = []
def gen_ply_file(i):
    dir = dirs[i]
    sdf_files = glob(dir+'/*.sdf')
    dir_name = sdf_files[0].split('/')[-2]
    out_dir = osp.join(save_base, dir_name)
    os.makedirs(out_dir, exist_ok=True)
    for lig_idx in range(len(sdf_files)):
        sdf_file = sdf_files[lig_idx]
        pkt_file = sdf_file[:-4] +'_pocket10.pdb'
        out_name = sdf_file.split('/')[-1][:-4]
        ply_file = "%s/%s_pocket_%s_res_%s.ply"%(out_dir, out_name, 8.0, 1.5)
        if os.path.isfile(ply_file):
            return 
        try:
            compute_inp_surface(pkt_file, sdf_file, out_dir, out_name=out_name, mesh_res=1.5,dist_threshold=8.0)
            return
        except: 
            print('Generate featured surface failed!{}'.format(i))
            fail_list.append(i)
            return i
```

```python
from tqdm import tqdm
fail_list = []
pool = multiprocessing.Pool(10)
for result in tqdm(pool.imap(gen_ply_file, range(len(dirs))), total=len(dirs)):
    if result is not None:
        fail_list.append(result)
pool.close()
pool.join()
```
